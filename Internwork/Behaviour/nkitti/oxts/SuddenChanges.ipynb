```python
def imp(d,*args):
    x=args[0]
    if(x==-1):
        try:
            return int(d)
        except:
            return d
    else:
        try:
            return float(d)
        except:
            return d
```

```python
for i in d.columns:
    x=(d[i][0].find('.'),0)
    #print(i,x)
    d[i]=d[i].apply(imp,args=x)
```

```python
ma=argrelextrema(np.array(d['vf']), np.greater)
ma=ma[0]
ma
```

```python
mi=argrelextrema(np.array(d['vf']), np.less)
mi=mi[0]
mi
```

```python
a=[]
b=[]
for i in range(len(ma)-1):
    x=-1*(d['vf'][mi[i]]-d['vf'][ma[i]])*100
    x1=((d['vf'][mi[i]]-d['vf'][ma[i]])/(mi[i]-ma[i]))*1000
    y=-1*(d['vf'][ma[i]]-d['vf'][mi[i+1]])*100
    y1=((d['vf'][ma[i]]-d['vf'][mi[i+1]])/(ma[i]-mi[i+1]))*1000
    a.append([round(x,4),x1,mi[i],ma[i]])
    b.append([round(y,4),y1,ma[i],mi[i+1]])
    #print("Incs: "+str(round(x1,4)),str(mi[i]),str(ma[i]))
    #print("Decs: "+str(round(y1,4)),str(ma[i]),str(mi[i+1]))
```

```python
a=sorted(a,key=lambda x:x[1],reverse=True)
b=sorted(b,key=lambda x:x[1],reverse=True)
```

```python
print("Incs")
print(a)
print("Decs")
print(b)
```
