```python
def imp(d,*args):
    x=args[0]
    if(x==-1):
        try:
            return int(d)
        except:
            return d
    else:
        try:
            return float(d)
        except:
            return d
```

```python
import math
def an(lat1,lo1):
    lat2=16.52638889
    lo2=80.64527778
    dlo=lo2-lo1
    y=math.sin(dlo)*math.cos(lat2)
    x=math.cos(lat1)*math.sin(lat2)-math.sin(lat1)*math.cos(lat2)*math.cos(dlo)
    z = math.atan2(y, x)
    z = math.degrees(z)
    z = (z + 360) % 360
    return z
```

```python
z=[]
for i in range(len(d)):
    z.append(an(d['field.latitude'][i],d['field.longitude'][i]))
```
